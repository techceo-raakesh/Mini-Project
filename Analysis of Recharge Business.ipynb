```python
df=df.dropna(how="all")
```

```python
df['Service Amt']=df['Service Amt'].fillna("Unknown")
```

```python
df.drop(['For','Mobile No./Cust.id'],axis=1,inplace=True)
```

```python
df['Earnings at EOD']=df["Earnings at EOD"].fillna("Unknown")
```

```python
df
```

```python
df=df.dropna(subset=['Date ','Particulars','Carrier','Payment Mode','R&T Status','Payment Status','Recharge'])
```

```python
df
```

```python
df.info()
```

```python
df['Recharge']=pd.to_numeric(df['Recharge'],errors='coerce')
```

```python
grouped=df.groupby('Carrier').sum().reset_index()
```

```python
plt.figure(figsize=(12,6)) 
c=['Red','Yellow','Purple', 'Orange', 'Teal']
plt.bar(grouped['Carrier'],grouped['Recharge'],color=c,edgecolor='black',linewidth=5,alpha=0.8)
plt.xlabel('Sim Brand')
plt.ylabel('Sum Of Recharge')
plt.title('Recharge of Sim',fontsize=14,fontweight='bold', color='darkblue')
plt.legend('Carrier')
plt.show()
```

```python
grouped=df.groupby('Payment Mode').sum().reset_index()
```

```python
plt.figure(figsize=(12,6)) 
c=['Red','Yellow','Purple', 'Orange', 'Teal']
plt.plot(grouped['Payment Mode'],grouped['Recharge'],linewidth=5,alpha=0.8)
plt.xlabel('Sim Brand')
plt.ylabel('Sum Of Recharge')
plt.title('Recharge of Sim',fontsize=14,fontweight='bold', color='darkblue')
for x, y in zip(grouped['Payment Mode'], grouped['Recharge']):
    plt.text(x, y + 30, str(y), ha='center', va='bottom', fontsize=10)

plt.legend(['Carrier'])
plt.show()
```
